```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
APT=pd.read_csv('APT.AX.csv')
EML=pd.read_csv('EML.AX.csv')
#OPY=pd.read_csv('OPY.AX.csv')
SPT=pd.read_csv('SPT.AX.csv')
Z1P=pd.read_csv('Z1P.AX.csv')
COMPANIES = [APT, EML, SPT, Z1P]
COMPANY_NAMES = ['APT', 'EML', 'SPT', 'Z1P']
```

```python
for company in COMPANIES:
    company['Date'] = pd.to_datetime(company['Date'], infer_datetime_format=True)
    company = company.set_index('Date')
```

# Let's create a graph, and normalise it so it is easier to read

```python
fig = go.Figure()
index=0
for company in COMPANIES:
    company_name = COMPANY_NAMES[index]
    fig.add_trace(go.Scatter(x=company['Date'], y=(company['Adj Close'])/(company['Adj Close'][0]),
                            mode='lines',
                            name=company_name))
    index+=1
fig.update_layout(template='plotly_dark')
fig.show()
```

#### BNPL did really well over this period:
    > Positive on a position taken and exited Feb 1; positive doubly so if you held through Coronavirus
    > Incredible returns post COVID

## How is the volume?

```python
fig = go.Figure()
index=0
for company in COMPANIES:
    company_name = COMPANY_NAMES[index]
    fig.add_trace(go.Scatter(x=company['Date'], y=(company['Volume']),
                            mode='lines',
                            name=str(company_name + ' Volu.')))
    index+=1
fig.update_layout(template='plotly_dark')
fig.show()
```

##### SPT volume on June 18 - 19 was due to a partnership announcement with Mastercard

```python
from statsmodels.tsa.stattools import adfuller
index=0
for company in COMPANIES:
    adf_test = adfuller(company['Close'])
    print('Company: ', COMPANY_NAMES[index])
    print('ADF Statistic: %f' % adf_test[0])
    print('p-value: %f' % adf_test[1])
    index+=1
```
